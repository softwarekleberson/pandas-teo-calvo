```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'nome': ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Alice'],
    'sobrenome': ['Smith', 'Jones', 'Brown', 'Calvo', 'Smith', 'Smith'],
    'salario': [70000, 80000, 90000, 60000, 70000, 75000]  
})
```

```python
df
```

```python
# Remover duplicatas com base na coluna 'nome e sobrenome'
df.drop_duplicates()
```

```python
# Remover duplicatas, mantendo a última ocorrência
df.drop_duplicates(keep='last')
```

```python
# Remover duplicatas com base em múltiplas colunas, sem ela não haveria mudança pois o salario é diferente
df = df.sort_values('salario', ascending=False)
df.drop_duplicates(keep='last', subset=['nome', 'sobrenome'])
```
