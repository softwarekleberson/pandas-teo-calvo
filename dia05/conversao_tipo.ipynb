```python
import pandas as pd
```

```python
df = pd.read_csv('../data/clientes.csv', sep=';')
df.head()
```

```python
# Ver a coluna qtdePontos e o seu tipo de dado
df['qtdePontos']
```

```python
# Alterar o tipo de dado da coluna qtdePontos para string
df['qtdePontos'].astype(str)
```

```python
# Alterar o tipo de dado da coluna qtdePontos para string e depois float
df['qtdePontos'].astype(str).astype(float)
```

```python
# Fazer a substituição dos valores "0000-00-00 00:00:00" na coluna DtCriacao por "2000-01-01 00:00:00", realizando a alteração in place
df['DtCriacao'] = df['DtCriacao'].replace({
    "0000-00-00 00:00:00": "2000-01-01 00:00:00"
})
```

```python
# Alterar o tipo de dado da coluna DtCriacao de string para datetime
pd.to_datetime(df['DtCriacao'])
```
