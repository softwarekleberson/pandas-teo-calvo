```python
import pandas as pd
```

```python
clientes = pd.read_csv('../data/clientes.csv', sep=';')
clientes
```

```python
# Removendo linhas com valores NA do DataFrame), se tiver algum valor NA em qualquer coluna
clientes.dropna()
```

```python
# Removendo linhas com valores NA do DataFrame, se todos os valores forem NA na linha
clientes.dropna(how='all')  
```

```python
df = pd.DataFrame({
    'nome': ["teo", None, "Nah", "Marcio"],
    'idade': [None, None, 43, 52],
    'salario': [3452, 4400, None, 5482]
})
```

```python
# Removendo linhas com valores NA na coluna "idade"
# Se passar salario no subset, tanto salario quanto idade devem estar com NA para a linha ser removida
df.dropna(how='all', subset=["idade", "salario"])
```

```python
# Se passar salario no subset, tanto salario quanto idade podem ser NA para a linha ser removida
# Desta vez se qualquer um dos dois for NA a linha será removida
df.dropna(how='any', subset=["idade", "salario"])
```

```python
# Preenchendo valores NA na coluna 'idade' com 0
# Caso queira trocar as series que tem NA, qualquer uma delas pode usar df.fillna(0)
df['idade'] = df['idade'].fillna(0)
```

```python
medias = df[['salario', 'idade']].mean()
medias
df.fillna(medias)
```
