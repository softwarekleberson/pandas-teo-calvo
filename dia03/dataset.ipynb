```python
df = pd.read_csv('../data/transacoes.csv', sep=';')
```

```python
df.head()
```

```python
# Filtrar linhas onde a coluna 'QtdePontos' é maior ou igual a 50 e menor que 100
filtro = (df['QtdePontos'] >= 50) & (df['QtdePontos'] < 100) & (df['DtCriacao'] > '2025-09-17')
df[filtro]
```

```python
# Filtrar linhas onde a coluna 'QtdePontos' é igual a 1 ou igual a 100
filtro = (df['QtdePontos'] == 1) | (df['QtdePontos'] == 100)
df[filtro]
```
