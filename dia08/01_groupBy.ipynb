```python
import pandas as pd
```

```python
cliente =  pd.read_csv('../data/clientes.csv', sep=';')
cliente.head()
```

```python
## Somando valores de uma coluna específica, no caso essa coluna apresenta valores 0 e 1
cliente["flTwitch"].sum()
```

```python
## Soma todos os valores da coluna.
## Divide o resultado pela quantidade total de elementos.
## Ou seja, soma todos os valores 1 e divide pela quantidade total de elementos.
cliente["flTwitch"].mean()
```

```python
## Saber a media de todas as redes sociais de uma vez só
redes_sociais = ["flEmail",	"flTwitch",	"flYouTube", "flBlueSky", "flInstagram"]
cliente[redes_sociais].mean()
```

```python
## Cria filtro para identificar colunas do tipo object
filtro = cliente.dtypes == "object"

## Retorna lista com nomes das colunas do tipo object
## Pois o filtro retorna uma Series booleana, onde True indica que a coluna é do tipo object, ao negar retorna tudo que não é object
colunas_numericas = cliente.dtypes[~filtro].index.tolist()
cliente[colunas_numericas].mean()
```

```python
## O describe retornar um dataframe, o mean, sum retorna uma serie 
cliente[colunas_numericas].describe()
```
