```python
import pandas as pd
```

```python
transacoes = pd.read_csv('../data/transacoes.csv', sep=';')
transacoes.head()
```

```python
# Ordenar o DataFrame pela coluna 'DtCriacao' em ordem decrescente
transacoes = transacoes.sort_values('DtCriacao', ascending=False)
transacoes['Data'] = pd.to_datetime(transacoes['DtCriacao']).dt.date
transacoes.drop_duplicates(keep='first', subset=['IdTransacao', 'Data'])
```
